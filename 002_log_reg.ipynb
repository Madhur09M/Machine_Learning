```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("cust_satisfaction.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df["Customer Type"].value_counts()
```

```python
loyal_customer = df[df["Customer Type"] == "Loyal Customer"]
disloyal_customer = df[df["Customer Type"] == "disloyal Customer"]
```

```python
loyal_customer = loyal_customer.sample(20000)
loyal_customer.shape
```

```python
balance_df = pd.concat([loyal_customer, disloyal_customer], axis = 0)
balance_df.shape
```

```python
cat_col = balance_df.select_dtypes(include = ["object"])
cat_col.head()
```

```python
num_col = df.select_dtypes(exclude = ["object"])
num_col.head()
```

```python
# cat_col["Gender"].value_counts()
```

```python
# pd.get_dummies(cat_col["Gender"]).astype(int).head()
pd.get_dummies(cat_col, drop_first = True).astype(int).head()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

one = OneHotEncoder(drop = "if_binary")
cat_col_encoded = one.fit_transform(cat_col).toarray()
cat_col_encoded
```

```python
column_name = list(one.get_feature_names_out())
column_name
```

```python
one_hot = pd.DataFrame(cat_col_encoded, columns = column_name)
one_hot.head()
```

```python
one_hot = one_hot.reset_index(drop = True)
num_col = num_col.reset_index(drop = True)
final_df = pd.concat([one_hot, num_col], axis = 1)
final_df.head()
```

```python
# dividing the data into dependent and independent variables

x = final_df.drop("satisfaction_satisfied", axis = 1)
y = final_df["satisfaction_satisfied"]
```

```python
# dividing the data into training and testing data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

## needs attention!!! ##

```python
x_train = x_train.dropna()
y_train = y_train.loc[x_train.index]

x_test = x_test.dropna()
y_test = y_test.loc[x_test.index]
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()
lgr.fit(x_train, y_train)
```

```python
lgr.score(x_train, y_train)
```

```python
lgr.score(x_test, y_test)
```

```python
y_pred = lgr.predict(x_test)
y_pred
```
